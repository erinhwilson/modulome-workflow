```python
def get_gene_imod_counts(ica_obj):
    b = ica_obj.M_binarized
    gene_imod_count = []
    for i, row in b.iterrows():
        c = sum(row.values)
        gene_imod_count.append((i,c))

    return dict(gene_imod_count)
```

```python
def view_imod_weights_on_genome(ica_data, imod_id):
    genes_in_imod = set(ica_data.view_imodulon(imod_id).index)
    gene2weight = dict(ica_data.M[imod_id].reset_index().values)
    temp_df = ica_data.gene_table[['gene_symbol','product','start_coord']].fillna('')
    temp_df.index.name='locus_tag'
    temp_df = temp_df.reset_index()
    temp_df['weight'] = temp_df['locus_tag'].apply(lambda x: gene2weight[x])
    temp_df['in_imod?'] = temp_df['locus_tag'].apply(lambda x: x in genes_in_imod)
    
    

    chart = alt.Chart(temp_df).mark_circle(size=60).encode(
        x=alt.X('start_coord:Q',axis=alt.Axis(title='Genomic Coordinate')),
        y=alt.Y('weight:Q',axis=alt.Axis(title=f'M weight for iModulon {imod_id}')),
        color='in_imod?:N',
        tooltip=['locus_tag:N', 'gene_symbol:N', 'product:N','in_imod?:N']
    ).interactive()
    return chart
```

```python
def view_imod_weights_dist(ica_data):
    top5_weights = []
    no_weights = []
    for i, row in ica_data.M.iterrows():
        weights = [np.abs(x) for x in row]
        weights = sorted(weights,reverse=True)
        if weights[0] <0.01:
            print(i, weights[:5])
        if weights[0] == 0.0:
            print(weights[:5])
            no_weights.append(i)
        else:
            top5_weights.append(weights[:5])

    cols = ['1st largest weight','2nd largest weight','3rd largest weight','4th largest weight','5th largest weight']
    top5_df = pd.DataFrame(top5_weights,columns=cols)
    
    for c in cols[:4]:
        plt.hist(top5_df[c].values, bins=20,label=c,alpha=0.2)
    plt.yscale('log')
    plt.legend()
    plt.xlabel("M weight")
    plt.ylabel("number of genes")
    plt.show()

    return no_weights
```

```python
def get_im_genes(ica_obj):
    im2g = {}
    for i in ica_obj.imodulon_names:
        im_genes = set(ica_obj.view_imodulon(i).index.values)
        im2g[i]=im_genes
    return im2g
```

```python
im2g = get_im_genes(ica_data)
im2g_og = get_im_genes(ica_data_og)
```

```python
# genes with no iModulon
im2g[-1] = set([x for x in gene_imod_dict if gene_imod_dict[x]==0.0])
im2g_og[-1] = set([x for x in gene_imod_dict_og if gene_imod_dict_og[x]==0.0])
```

```python
# calculate the overlap similarity 
def overlap_sim(s1, s2):
    n = len(s1.intersection(s2))
    d = min(len(s1),len(s2))
    if d == 0:
        return 0
    return n/d

def jacc_sim(s1,s2):
    n = len(s1.intersection(s2))
    d = len(s1 | s2)
    if d == 0:
        return 0 
    return n/d

def get_sim(map1, map2,map1_name, map2_name):
    data = []
    for s1 in map1:
        for s2 in map2:
            o = overlap_sim(map1[s1], map2[s2])
            j = jacc_sim(map1[s1], map2[s2])
            s1_size = len(map1[s1])
            s2_size = len(map2[s2])
            data.append((s1,s2,o,j,s1_size,s2_size))

    return pd.DataFrame(data, columns=[map1_name,map2_name,'osim','jsim',f'{map1_name}_size', f'{map2_name}_size'])

```

```python
# how much do these iModulon calls overlap?
sims = get_sim(im2g_og, im2g,'iMod_lowExAdj','iMod_20221031')
```

```python
im2ev = dict(ica_data.imodulon_table.reset_index()[['index','explained_variance']].values)
im2ev_og = dict(ica_data_og.imodulon_table.reset_index()[['index','explained_variance']].values)

sims['iMod_lowExAdj_exp_var'] = sims['iMod_lowExAdj'].apply(lambda x: im2ev_og[x] if x in im2ev_og else 0)
sims['iMod_20221031_exp_var'] = sims['iMod_20221031'].apply(lambda x: im2ev[x] if x in im2ev else 0)

sims
```

```python
sims_sort = sims.sort_values(["iMod_lowExAdj_exp_var",'iMod_20221031_exp_var'],ascending=[False,False])

sorted_heat_chart = alt.Chart(sims_sort).mark_rect().encode(
    x=alt.X('iMod_lowExAdj:O',sort=alt.EncodingSortField(field='iMod_lowExAdj_exp_var:O', order='descending')),
    y=alt.Y('iMod_20221031:O',sort=alt.EncodingSortField(field='iMod_20221031_exp_var:O', order='descending')),
    color='jsim:Q',
    tooltip=['iMod_20221031:O','iMod_20221031_size:Q','iMod_lowExAdj:O','iMod_lowExAdj_size:Q','jsim:Q','osim:Q']
)
sorted_heat_chart
```
